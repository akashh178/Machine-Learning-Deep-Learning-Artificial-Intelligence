```python
import numpy as np
import matplotlib.pyplot as plt
```

data distribution refers to the way data points are spread or organized within a dataset. It describes the frequency or occurrence of different values in your data, such as how many times a particular value or range of values appears.

Normal Distribution <br> The normal distribution is a continuous probability distribution that is symmetric around the mean.<br> Most values cluster around the mean, and the probabilities taper off equally in both directions from the mean. <br>The classic "bell curve" shape is characteristic of normal distributions.

```python
mean = 0
std_dev = 1
data = np.random.normal(mean, std_dev, 1000)
```

```python
plt.hist(data, bins = 30)
```

Binomial Distribution <br>
The binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, either Success or Failure.


Example: Simulate flipping a coin 10 times and observe how many heads you get (with a 50% probability of getting heads).

```python
n_trials = 10 #Number of trials
p_success = 0.5 #probability of success

#generating 1000 samples of 10 coin flips
binomial_data = np.random.binomial(n = n_trials, p = p_success, size = 1000)
```

```python
plt.hist(binomial_data, bins = 10)
plt.title('Binomial distribution (n=10, p = 0.5)')
plt.show()
```

Logistic Distribution <br>
The logistic distribution is similar to the normal distribution but has wider tails than a normal distribution, which makes it more consistent with underlying data and better able to provide insight into the likelihood of extreme events.

```python
mean = 0
scale = 1

logistic_data = np.random.logistic(mean, scale, 1000)
```

```python
plt.hist(logistic_data, bins=30)
plt.show()
```

exponential distribution <br>
the exponential distribution is a continuous probability distribution that often concerns the amount of time until some specific event happens. It is a process in which events happen continuously and independently at a constant average rate.

Memoryless Property: One of the unique properties of the exponential distribution is that it is memoryless. This means that the probability of an event occurring in the next ( t ) units of time is the same, regardless of how much time has already elapsed. <br><br>
The probability of an event occurring in the future is independent of how long it's been since the last event. For example, if you're waiting for a customer to arrive, the distribution of how long you'll wait doesn't change based on how long you've already been waiting. <br><br>
Parameter :
The exponential distribution has one parameter, Î», which represents the average rate of events in a given time interval.

Uses :
The exponential distribution is used to model things like the time between customer arrivals at a store, the time it takes a copy machine to work between repairs, or the time until a TV system fails. <br><br>
Skewed shape :
The exponential distribution is skewed, with a vertical rise at zero and a gradual descent with a long tail on the right.<br><br>

Rate Parameter (( \lambda )):
( \lambda ) is the rate at which events occur. For example, if ( \lambda = 2 ), it means that, on average, 2 events occur per unit of time. <br><br>
Mean Time Between Events:
The mean (expected value) of the exponential distribution is ( \frac{1}{\lambda} ). This is because if events occur at a rate of ( \lambda ) per unit time, the average time between consecutive events is ( \frac{1}{\lambda} ) units of time.
For instance, if ( \lambda = 2 ), the mean time between events is ( \frac{1}{2} = 0.5 ) units of time.

```python
#Example: Model the time between events happening at an average rate of 1 event per unit time.

lambda_param = 1.0

exp_data = np.random.exponential(1/lambda_param, 1000)
```

```python
plt.hist(exp_data, bins=30)
plt.show()
```

Multinomial Distribution :
The multinomial distribution generalizes the binomial distribution. Instead of two outcomes (success/failure), you have multiple outcomes (k outcomes) in a fixed number of trials.

```python
#Example: Roll a die 10 times and see the frequency of each face (1-6).

n_trials = 10
pvals = [1/6]*6 # Probabilities for each outcome

multinomial_data = np.random.multinomial(n_trials, pvals, size = 1000)

# Summing results over 1000 simulations
sum_of_rolls = np.sum(multinomial_data, axis = 0)
```

```python
print(multinomial_data)
```

```python

sum_of_rolls
```

```python
plt.bar(range(1, 7), sum_of_rolls / sum(sum_of_rolls))
plt.title('Multinomial Distribution (Fair Die Roll)')
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.show()
```
