```python
from transformers import pipeline
```

```python
clf = pipeline("sentiment-analysis")
```

```python
clf('I will not go for byying milk')
```

```python
clf('he failed in driving test, but he is happy')
```

```python
clf = pipeline("zero-shot-classification")

clf(
    "This notebook is about NLP",
    candidate_labels = ["education", "politics", "business"]
)
```

```python
gen = pipeline("text-generation")
```

```python

gen("this notebook is made up of")
```

```python
unmasker = pipeline("fill-mask")
```

```python
unmasker("this chair is so heavy , it is made of <mask>.", top_k = 5)
```

```python
ner = pipeline("ner", grouped_entities = True,)
```

```python
ner("i am akash, i study in Ahmedabad")
```

```python
qa = pipeline("question-answering")
```

```python
qa(
    question = "where do i live?",
    context = "my name is akash, i live in ahmedabad"
)
```

```python
summarizer = pipeline("summarization")
```

```python
summarizer(
    '''As we know, For building the machine learning model we need data related
     to our problem statements, Sometimes we have our data and Sometimes we
     have to find it. Text data is available on websites, in emails, in social
     media, in form of pdf, and many more. But the challenge is. Is it in a
     machine-readable format? if in the machine-readable format then will it
     be relevant to our problem? So, First thing we need to understand our problem
      or task then we should search for data. Here we will see some of the ways
       of collecting data if it is not available in our
       local machine or database.''', max_length = 100, min_length = 64
)
```
